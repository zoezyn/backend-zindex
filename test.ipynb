```python
# Open and read the file
with open("/Users/zoe.yan/Desktop/temp/My Clippings.txt", "r") as file:
    content = file.read()  # Reads the entire file
print(content)
```

```python
def process_kindle_notes(content: str) -> dict:
    books = {}
    sections = content.split('==========')
    
    for section in sections:
        if section.strip():
            lines = section.strip().split('\n\n')
            # print(len(lines) if len(lines) > 1 else "no notes")
            if len(lines) > 1:
                title_location = lines[0].strip().split('\n')
                book_notes = lines[1]
                # print(type(book_notes))


                book_title = title_location[0]
                note_location = title_location[1]
   
                if book_title not in books:
                    books[book_title] = []
                    
                if book_notes not in books[book_title]:
                    books[book_title].append((book_notes, note_location))    
    return books

books = process_kindle_notes(content)
books
```
